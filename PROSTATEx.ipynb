We used PROSTATEx cases for analysis. There are two spreadsheets regarding PROSTATEx.

```D```: parameters of the scanner in this file

```AWR```: in-house prostate MRI scoring in this file

Our goal: to include 150 PROSTATEx studies in our dataset

````python
with open("ProstateX_MR_DICOM_Info_DAK.csv", "r", encoding="utf-8") as file:
    D = pd.read_csv(file)

AWR = pd.read_csv('PROSTATEx_patients_AWR.csv')
AWR = AWR.rename(columns = {'Patient': 'PatientID'})
````

````python
display(D)
display(AWR)
````

Summary of the scanners' parameters in PROSTATEx

````python
D["AcquisitionDate"] = pd.to_datetime(D["AcquisitionDate"], format='%Y%m%d')
````

We are not interested in the scanners' parameters except for the models of the scanners in ```D```. Hence, I grouped and cleaned the dataframe ```D``` to figure out the scanner used for each patient, and assigned the new dataframe as ```D1```

````python
D1 = D[['PatientID','Model']].drop_duplicates()
D1 = D1.reset_index()
D1
````



```D1``` and ```AWR``` were inner-joined by using ```PatientID``` as a key


````python
Px = pd.merge(D1, AWR, how = 'inner', on = 'PatientID')
Px
````


Since we wanted to set up a heterogeneous/balanced dataset, and only 57 patients were scanned by ```TrioTim```, so we deliberately included all patients scanner by ```TrioTim```, and the following lines will work out how to filter patients scanned by ```Skyra```


Check how T2WI and DWI were scanned in each patient


Select the columns

```DWI Q lo-hi```: in-house quality score of DWI, is a five-point score 


```T2 Q lo-hi```: in-house quality score of T2WI, is a five-point score

```Score``` = ```DWI Q lo-hi``` + ```T2 Q lo-hi```, is a ten-point score


````python
Px = Px[['Study_Day', 'PatientID', 'Model', 'Series_DWI_Desc', 'DWI Q lo-hi', 'T2 Q lo-hi']]
Px['Score'] = Px['DWI Q lo-hi'] + Px['T2 Q lo-hi']
Px = Px.rename(columns = {'Study_Day': 'Date.MRI'})
Px = Px.rename(columns = {'PatientID': 'PID'})
Px
````


Since we wanted to set up a heterogeneous dataset, and only 57 patients were scanned by ```TrioTim```, so we deliberately included all patients scanner by ```TrioTim```. For patients not scanned by ```TrioTim```, first, we looked at the distribution of ```Score```


````python
Px[Px['Model'] != "TrioTim"]['Score'].value_counts().sort_index(ascending=False)
````


Most of the MRI studies' quality is good (```Score``` is ```8-9```), and we wanted to collect a balanced dataset, so we intendedly included all the studies whose score is ```<8``` or ```>10```, and saved it as ```out``` dataframe


````python
model_not_TrioTim = Px[Px['Model'] != "TrioTim"]
model_not_TrioTim 
out = model_not_TrioTim[(model_not_TrioTim['Score'] < 8) | (model_not_TrioTim['Score'] ==10)]
out = out.reset_index(drop = True)
out
````


Append patients scanned by ```TrioTim``` to ```out```


````python
temp = Px[Px['Model'] == "TrioTim"]
out = pd.concat([out, temp]).reset_index(drop=True)
````

````python
out
````


We have got 99 studies, and we wanted to include 150 studies. For the remaining cases (i.e. not scanned by ```TrioTim```, and the ```Score``` is ```8 or 9```, 51 cases were randomly selected into ```out```


````python
temp = model_not_TrioTim[(model_not_TrioTim['Score'] >= 8) & (model_not_TrioTim['Score'] < 10)]
temp 
````

````python
temp = temp.sample(n=51, random_state=41)
````

````python
out = pd.concat([out, temp]).reset_index(drop=True)
out
````


Finally, we collected 150 PROSTATEx cases with a (somehow) balanced manner
